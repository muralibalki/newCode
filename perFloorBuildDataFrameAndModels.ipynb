```python
import numpy as np
import pandas as pd
from scipy.cluster.vq import *
import operator
from matplotlib import pyplot as plt   
import os
from subprocess import Popen, PIPE
%pylab inline
# some display options to make figures bigger
# hide this
pd.set_option('display.max_columns', 15)
pd.set_option('display.line_width', 400)
pd.set_option('display.mpl_style', 'default')
rcParams['figure.figsize'] = (14, 7)
import matplotlib
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
df = pd.read_csv('/home/murali/building/data/Outside Weather_Temperature.csv', index_col=False, header=1,low_memory=False);
```

```python
df.columns=['TimeStamp','Temperature']
df['TimeStamp']=pd.to_datetime(df['TimeStamp'])
df=df.set_index(['TimeStamp'])
df=df.Temperature.resample('5min')
df=pd.DataFrame(list(df.values), index=df.index)
df.columns=['Temperature']
oldDf=df;
```

```python
sensorNames=['Actual Cooling Setpoint_PresentValue','Actual Heating Setpoint_PresentValue','Zone Temperature_PresentValue','Actual Supply Flow_PresentValue','Occupied Command_PresentValue','Damper Position_PresentValue','Supply Vel Press_PresentValue','WarmCool Adjust_PresentValue','Cooling Max Flow_PresentValue']
```

```python
cmd="ls -D1 /home/murali/building/data/" + FLOOR + "*"
a=Popen(cmd, shell=True, stdout=PIPE).communicate()[0]
b=a.split('\n')
nSensorsTot=len(b)
listOfRooms=b
for i in xrange(0,nSensorsTot-1):
    listOfRooms[i]=b[i].split('/')[5][0:4]
del listOfRooms[-1]
listOfRooms=list(set(listOfRooms))
```

```python
count=0
flag=1
while count < len(listOfRooms):
    room=str(listOfRooms[count])
    print room
    count=count+1
    for sensor in sensorNames:
        df2=[];
        fileName='/home/murali/building/data/'+room+'_'+sensor+'.csv'
        try:
            df2 = pd.read_csv(fileName, index_col=False, header=1,low_memory=False);
            df2.columns=['TimeStamp','DP']
            df2['TimeStamp']=pd.to_datetime(df2['TimeStamp'])
            df2=df2.set_index(['TimeStamp'])
            df2=df2.DP.resample('5min')
            df2=pd.DataFrame(list(df2.values), index=df2.index)
            df2=df2.fillna(method='pad')
            df2=df2.reset_index()
            olddf2=df2
        except IOError:
            print "Missing file"+fileName     
            df2=olddf2

        df2['sensor']=sensor
        df2['room']=room
        df2=pd.DataFrame(matrix(df2))
        df2.columns=['TimeStamp','Value','sensor','room']
        cols = [c for c in df2.columns if c not in {'TimeStamp', 'Value'}]
        df2 = df2.set_index(cols + ['TimeStamp']).unstack('TimeStamp')
        df2.columns = df2.columns.levels[1]
        df2.columns.name = None

        if flag:
            df=df2
            flag=0
        else:
            df=df.append(df2)
        #wait = input("PRESS ENTER TO CONTINUE.")
```

```python
df.to_pickle('SecondFloorLarge.pkl')
```
